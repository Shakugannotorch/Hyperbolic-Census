```python
import regina
```

### $n=9$

```python
n=9

sp_count = 0
tk_count = 0 
lf_count = 0 
done = False

with open("../data/tricensus_"+str(n)+"/leftovers/level_2.dat","r") as f:
    with open("../data/tricensus_"+str(n)+"/leftovers/level_3.dat","w") as lf:
        with open("../data/tricensus_"+str(n)+"/ruled_out/plane_found.dat","w") as ro:
            for sg in f:
                done = False
                t = regina.Triangulation3(sg.strip())
                nss = regina.NormalSurfaces(t,regina.NS_STANDARD)

                for ns in nss:
                    if ns.eulerChar()>0:
                        sp_count+=1
                        ro.write(sg)
                        done = True
                        break
                
                if not done:
                    for ns in nss:
                        if ns.eulerChar() == 0 and not ns.isVertexLink():

                            if not ns.isTwoSided():
                                tk_count +=1
                                ro.write(sg)
                                done = True
                                break
                            else:
                                cut_t = ns.cutAlong()
                                cut_t.intelligentSimplify()
                                found = True

                                for comp in cut_t.triangulateComponents():

                                    if not comp.isOrientable():
                                        comp.makeDoubleCover()
                                    assert comp.isOrientable()
                                    
                                    comp.intelligentSimplify()
                                    comp.knowsSolidTorus()

                                    if comp.isSolidTorus(): 
                                        found = False
                                        break
                                
                                if found:
                                    tk_count += 1
                                    ro.write(sg)
                                    done = True
                                    break

                if not done:
                    lf_count += 1
                    lf.write(sg)


print("spheres and projective planes found: " + str(sp_count))
print("Tori and Klein bottles found: " + str(tk_count))
print("leave out: " + str(lf_count))
```
